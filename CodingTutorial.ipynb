### VS Code

##### Big Ideas:

Workspace: A folder that contains all of the files for a project.

 - You can open a workspace by opening the folder in VS Code.

Integrated Terminal: A terminal that is built into VS Code.

 - You can open the integrated terminal by pressing `Ctrl + ~` or by clicking `View > Terminal` in the menu bar.

Advanced Coding Functionality: VS Code has a lot of advanced functionality

 - You can right click on variable and function names to do a number of things, including renaming them. This is called "refactoring".
    - You can right click on a variable and click "Go to Definition" to go to the place where that variable is defined.
    - You can right click on a variable and click "Peek Definition" to see the definition of that variable without leaving the file you are in.
    - You can right click on a variable and click "Find All References" to see all of the places where that variable is used.
    - You can right click on a variable and click "Rename Symbol" to rename that variable everywhere it is used.
    - You can right click on a variable and click "Change All Occurrences" to change all occurrences of that variable to a different variable.

 - You can navigate and edit files more efficiently with keyboard shortcuts. 
    - My personal faves:
        - `Alt + Click` to place multiple cursors.
        - `Alt + Up/Down` to move a line up or down.
        - `Alt + Shift + Up/Down` to copy a line up or down.
        - `Alt + Left/Right` to go back or forward.
        - `Ctrl + Alt + Up/Down` to place multiple cursors above or below.
        - `Ctrl + Shift + K` to delete a line.
        - `Ctrl + /` to comment out a line.
    - You can find the shortcut reference by clicking `Help > Keyboard Shortcuts` in the menu bar.

Extensions: User-created mods(modifications) to VS Code that add functionality.

 - You can install extensions by clicking the "Extensions" icon in the left sidebar.
 - Some essentials (IMO):
    - Language Specific Extensions
        - Python, C++, etc.
        - Provides syntax highlighting and other language-specific features.
        - Colors your code, makes it easier to read, and points out errors.
    - Remote
        - Allows you to connect to a remote server(via ssh) and edit files on that server.
        - You don't have to only work in the terminal on an ssh machine(like ISSAC) you can use VS Code.
    - Themes
        - Any programmer will tell you that a dark theme is essential.

VS Code is a very powerful tool. It is worth your time to learn how to use it well.


### Bash

##### Big Ideas:

Commands: Bash is a command line interface. You can type commands into the terminal and it will execute them.

 - You can type `ls` to list the files in the current directory.
 - You can type `cd` to change directories.
 - You can type `pwd` to print the current working directory.
 - You can type `mkdir` to make a directory.
 - You can type `touch` to make a file.
 - You can type `rm` to remove a file.
 - You can type `rm -r` to remove a directory.
 - You can type `mv` to move a file or directory.
 - You can type `cp` to copy a file or directory.
 - You can type `cat` to print the contents of a file.
 - You can type `echo` to print a string.
 - You can type `>` to redirect the output of a command to a file.
 - You can type `>>` to append the output of a command to a file.
 - You can type `|` to pipe the output of a command to another command.
 - You can type `grep` to search for a string in a file.
 - You can type `man` to get the manual for a command.
 - You can type `history` to see a list of all of the commands you have typed.
 - You can type `!` to run a command from your history.
 - You can type `Ctrl + R` to search your history.
 - You can type `Ctrl + C` to kill a process.
 - You can type `Ctrl + Z` to pause a process.
 - You can type `fg` to resume a process.
 - You can type `bg` to run a process in the background.
 - You can type `jobs` to see a list of all of the processes you have running.
 - You can type `chmod` to change the permissions of a file or directory.
 - You can type `sudo` to run a command as the superuser.
 - You can type `alias` to create an alias for a command.
 - You can type `source` to run a bash script.

Environment Variables: bash is a program, just like the ones you write. It has variables that you can access and change.

 - You can type `env` to see a list of all of the environment variables.
 - You can type `echo $` to print the value of an environment variable.
 - You can type `export` to create an environment variable.
 - You can type `unset` to delete an environment variable.

 - Here are some of the most important environment variables:
    - `PATH`: A list of directories that bash will search for commands.
    - `HOME`: The home directory of the current user. You can type `cd ~` to go to the home directory.
    - `USER`: The username of the current user.
    - `EDITOR`: The default editor for bash to use.
    - `SHELL`: The default shell for bash to use.
    - `TERM`: The default terminal for bash to use.
    - `PS1`: The prompt that bash will use.
    - `PS2`: The prompt that bash will use when it is expecting more input.
    - `PS3`: The prompt that bash will use when it is expecting a choice.
    - `PS4`: The prompt that bash will use when it is tracing a command.

    - You can type `export PATH=$PATH:` to add a directory to the PATH.

Package Manager: The app store for ubuntu `apt`

 - If you want to install a new command you can use `apt`.
   - You can type `apt-get install` to install a package.
   - You can type `apt remove` to remove a package.
   - You can type `apt search` to search for a package.
   - You can type `apt show` to show information about a package.
   - You can type `apt list` to list all of the packages.
   - You can type `apt list --installed` to list all of the installed packages.

   For python packages you can use `pip`.
   - You can type `pip install` to install a package.
    - You can type `pip uninstall` to uninstall a package.
